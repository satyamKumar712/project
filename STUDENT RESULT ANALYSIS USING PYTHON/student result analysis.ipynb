```python
df["TOTAL"] = df["Maths"] + df["Science"] + df["English"]
print(df.head())
```

```python
df["RESULT"] = "FAIL"
df.loc[df["TOTAL"] >= 180  , "RESULT"] = "PASS"
print(df.head())
```

```python
print(df.columns)
```

```python
df["AVERAGE"] = df["TOTAL"]/3

print(df.head())
```

```python
print("top 10 student")
highest = df.nlargest(5 , "TOTAL")

print(highest)
```

```python
print("lowest 10 student")
lowest = df.nsmallest(5 , "TOTAL")

print(lowest)
```

```python
subject = ["Maths" , "Science" , "English"]
avg_marks = df[subject].mean()
plt.figure(figsize=(8,3))

#bar graph

plt.subplot(1,3,1)
plt.bar(subject , avg_marks)
plt.xlabel("subjects")
plt.ylabel("avg marks")

#scatter plot

plt.subplot(1,3,2)
plt.scatter(subject , avg_marks)
plt.xlabel("subjects")
plt.ylabel("avg marks")

# line graph

plt.subplot(1,3,3)
plt.plot(subject , avg_marks)
plt.xlabel("subjects")
plt.ylabel("avg marks")


plt.tight_layout()
plt.show()
```

```python
#top 10 student highest marks
plt.bar(highest["Name"] , highest["TOTAL"] , color ="green")
plt.title("TOP 10 STUDENT HIGHEST MARKS")

plt.xlabel("NAME")
plt.ylabel("MARKS")
plt.show()
```

```python
#top 10 student lowest marks
plt.bar(lowest["Name"] , lowest["TOTAL"] , color ="red")
plt.title("TOP 10 STUDENT HIGHEST MARKS")

plt.xlabel("NAME")
plt.ylabel("MARKS")
plt.show()
```

```python
#total marks distribution
plt.hist(df["TOTAL"] , bins = 10)
plt.title("TOTAL MARKS DISTRIBUTION")
plt.show()
```

```python
#math vs science comparison
plt.bar(df["Maths"] , df["Science"] , color = "blue")
plt.title("maths vs science")
plt.xlabel("maths")
plt.ylabel("science")
plt.show()
```

```python
#pie chart
result = df["RESULT"].value_counts()
plt.pie(result , labels = result.index , autopct="%1.1f%%")
plt.title("PASS VS FAIL")
plt.show()
```
