# COMPARISON GRAPH

```python
name = df.groupby("Name")["Salary"].sum().nlargest(5)
same = df.groupby("Name")["Salary"].sum().nsmallest(5)
mexp = df.groupby("Name")["Experience_Years"].sum().nlargest(5)
lexp = df.groupby("Name")["Experience_Years"].sum().nlargest(5)

plt.figure(figsize=(12,10))

plt.subplot(2,2,1)
plt.bar(name.index , name.values)
plt.title("NAME WISE HIGH SALARY")

plt.subplot(2,2,2)
plt.bar(same.index , same.values)
plt.title("NAME WISE LOW SALARY")

plt.subplot(2,2,3)
plt.bar(mexp.index , mexp.values)
plt.title("NAME WISE MAX EXP")

plt.subplot(2,2,4)
plt.plot(lexp.index , lexp.values)
plt.title("NAME WISE LOW EXP")


plt.show()
```

# COMPARISON GRAPH

```python
n = df.groupby("Name")["Age"].sum().nlargest(5)
e = df.groupby("Education")["Experience_Years"].sum().nlargest(10)
at = df.groupby("Name")["Attendance_%"].sum().nlargest(5)
gd =  df.groupby("Gender")["Performance_Rating"].sum().nlargest(10)
dt =  df.groupby("Department")["Job_Role"].count().nlargest(5)
mt = df["Marital_Status"].value_counts()

plt.figure(figsize = (16,10))

plt.subplot(2,3,1)
plt.bar(n.index , n.values , color = "red")
plt.title("NAME WISE HIGH AGE")

plt.subplot(2,3,2)
plt.bar(e.index , e.values , color = "red")
plt.title("EDU WISE HIGH EXP")

plt.subplot(2,3,3)
plt.bar(at.index , at.values , color = "red")
plt.title("NAME WISE HIGH ATTENDANCE")

plt.subplot(2,3,4)
plt.plot(gd.index , gd.values , color = "red")
plt.title("GENDER WISE HIGH PERFO. RATING")

plt.subplot(2,3,5)
plt.bar(dt.index , dt.values , color = "red")
plt.title("DEPT.  WISE HIGH JOB ROLE")

plt.subplot(2,3,6)
plt.pie(mt.values , labels = mt.index , autopct = "%1.1f%%" )
plt.title("MARITAL STA.")
plt.show()
```
