# top 5 product wise total sales

```python

sales = df.groupby("Product")["Sales"].sum().nlargest(5)
plt.bar(sales.index, sales.values)
plt.title("TOP 5  SELLING PRODUCT")
plt.xlabel("PRODUCT")
plt.ylabel("SALES")
plt.show()
```

# Quantity sold by product

```python

plt.figure(figsize =(10 ,12))
plt.subplot(1,2,1)
qty = df.groupby("Product")["Quantity"].sum().nlargest(5)
plt.bar(qty.index , qty.values)
plt.xlabel("PRODUCT")
plt.ylabel("QUANTITY")

plt.subplot(1,2,2)
qty = df.groupby("Product")["Quantity"].sum().nsmallest(5)
plt.bar(qty.index , qty.values)
plt.xlabel("PRODUCT")
plt.ylabel("QUANTITY")
plt.show()
```

```python

state = df.groupby("State")["Sales"].sum().nlargest(5)
plt.pie(state.values , labels = state.index , autopct = "%1.1f%%")
plt.title("TOP 5 STATE WISE SALES")
plt.show()        
        
```

```python
category = df.groupby("Category")["Sales"].sum().nlargest(3)
plt.pie(category.values , labels = category.index , autopct = "%1.1f%%")
plt.title("TOP 5 CATEGORY WISE SALES")
plt.show()        
        
```

```python
quantity = df.groupby("Product")["Quantity"].sum().nlargest(5)
plt.pie(quantity.values , labels = quantity.index , autopct = "%1.1f%%")
plt.title("TOP 5 QUANTITY WISE PRODUCT")
plt.show()        
        
```

```python
profit = df.groupby("Category")["Profit"].sum().nlargest(3)
plt.pie(profit.values , labels = profit.index , autopct = "%1.1f%%")
plt.title("TOP 3 CATEGORY WISE PROFIT")
plt.show()        
        
```

# comparison graph

```python

product_sales = df.groupby("Product")["Sales"].sum()
category_sales = df.groupby("Category")["Sales"].sum()
region_sales = df.groupby("State")["Sales"].sum()
product_quantity = df.groupby("Product")["Quantity"].sum()

# Figure
plt.figure(figsize=(12,10))

# Graph 1
plt.subplot(2,2,1)
plt.bar(product_sales.index, product_sales.values)
plt.title("Product Wise Sales")
plt.xticks(rotation=45)

# Graph 2
plt.subplot(2,2,2)
plt.bar(category_sales.index, category_sales.values)
plt.title("Category Wise Sales")
plt.xticks(rotation=45)

# Graph 3
plt.subplot(2,2,3)
plt.pie(region_sales.values, labels=region_sales.index, autopct="%1.1f%%")
plt.title("state Wise Sales")

# Graph 4
plt.subplot(2,2,4)
plt.plot(product_quantity.index, product_quantity.values, marker="o")
plt.title("Product Wise Quantity")
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

# comparison graph

```python

product_sales = df.groupby("Product")["Sales"].sum()
category_sales = df.groupby("Category")["Sales"].sum()
region_sales = df.groupby("State")["Sales"].sum()
product_quantity = df.groupby("Product")["Quantity"].sum()
category_profit = df.groupby("Category")["Profit"].sum()
top5 = df.groupby("Product")["Sales"].sum().nlargest(5)

# Figure
plt.figure(figsize=(16,10))

# 1. Product Wise Sales
plt.subplot(2,3,1)
plt.bar(product_sales.index, product_sales.values)
plt.title("Product Wise Sales")
plt.xticks(rotation=45)

# 2. Category Wise Sales
plt.subplot(2,3,2)
plt.bar(category_sales.index, category_sales.values)
plt.title("Category Wise Sales")

# 3. Region Wise Sales
plt.subplot(2,3,3)
plt.pie(region_sales.values, labels=region_sales.index, autopct="%1.1f%%")
plt.title("State Wise Sales")

# 4. Product Wise Quantity
plt.subplot(2,3,4)
plt.plot(product_quantity.index, product_quantity.values, marker='o')
plt.title("Product Wise Quantity")
plt.xticks(rotation=45)

# 5. Category Wise Profit
plt.subplot(2,3,5)
plt.bar(category_profit.index, category_profit.values)
plt.title("Category Wise Profit")

# 6. Top 5 Products
plt.subplot(2,3,6)
plt.bar(top5.index, top5.values)
plt.title("Top 5 Selling Products")
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
plt.savefig("comparison.png"
```
